```python
import torch

# === 1. Dosyayƒ± y√ºkle ===
ckpt_path = "/content/ckpts/MVTec AD/wood/BEST_P_PRO.pth"
checkpoint = torch.load(ckpt_path, map_location="cpu")

# === 2. Her bir bile≈üenin i√ßini yazdƒ±r ===
for key in ["tt", "bn", "st", "dfs"]:
    print(f"\nüîé ƒ∞√ßerik: {key.upper()} (Anahtar: '{key}')")

    if key not in checkpoint:
        print(f"‚ùå '{key}' bulunamadƒ±.")
        continue

    state_dict = checkpoint[key]

    for name, param in state_dict.items():
        print(f" - {name} : {param.shape}")
        # ƒ∞steƒüe baƒülƒ±: Aƒüƒ±rlƒ±k deƒüerinin ilk birka√ß elemanƒ±nƒ± g√∂ster
        print(f"   √∂rnek deƒüerler: {param.view(-1)[:5].tolist()}")  # ƒ∞lk 5 eleman
```
