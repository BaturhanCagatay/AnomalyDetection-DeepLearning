```python
import os
import sys
import cv2
import torch
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from tqdm import tqdm
from torchvision import transforms
from functools import partial
import torch.nn as nn

# INP-Former dizinini Python yoluna ekle
sys.path.append("/content/INP-Former")

# Modülleri yükle
from models import vit_encoder
from models.uad import INP_Former
from models.vision_transformer import Mlp, Aggregation_Block, Prototype_Block
from utils import get_gaussian_kernel, cal_anomaly_maps

# === DİZİN AYARLARI ===
defect_test_dir = "/content/drive/MyDrive/wood_dataset/wood/test/defect"
gt_dir = "/content/drive/MyDrive/wood_dataset/wood/ground_truth/defect"
# Eğitimde kullanılan parametrelerle aynı model yolu
model_path = "/content/inp_result/INP-Former-Single-Class_dataset=MVTec-AD_Encoder=dinov2reg_vit_base_14_Resize=252_Crop=252_INP_num=6/wood/model.pth"
output_dir = "/content/anomaly_groundtruth_comparisoNn"

# Output dizini oluştur
os.makedirs(output_dir, exist_ok=True)

# === MODEL YÜKLEME ===
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Cihaz: {device}")

# Eğitimde kullanılan parametreler
INP_num = 6  # Eğitimde kullanılan INP_num değeri
embed_dim = 768
num_heads = 12
target_layers = [2, 3, 4, 5, 6, 7, 8, 9]
fuse_layer_encoder = [[0, 1, 2, 3], [4, 5, 6, 7]]
fuse_layer_decoder = [[0, 1, 2, 3], [4, 5, 6, 7]]

# Model bileşenlerini oluştur
encoder = vit_encoder.load('dinov2reg_vit_base_14')
Bottleneck = nn.ModuleList([Mlp(embed_dim, embed_dim * 4, embed_dim, drop=0.)])
INP = nn.ParameterList([nn.Parameter(torch.randn(INP_num, embed_dim))])
INP_Extractor = nn.ModuleList([
    Aggregation_Block(dim=embed_dim, num_heads=num_heads, mlp_ratio=4., qkv_bias=True,
                     norm_layer=partial(nn.LayerNorm, eps=1e-8))
])
INP_Guided_Decoder = nn.ModuleList([
    Prototype_Block(dim=embed_dim, num_heads=num_heads, mlp_ratio=4., qkv_bias=True,
                   norm_layer=partial(nn.LayerNorm, eps=1e-8)) for _ in range(8)
])

# Modeli oluştur
model = INP_Former(
    encoder=encoder,
    bottleneck=Bottleneck,
    aggregation=INP_Extractor,
    decoder=INP_Guided_Decoder,
    target_layers=target_layers,
    remove_class_token=True,
    fuse_layer_encoder=fuse_layer_encoder,
    fuse_layer_decoder=fuse_layer_decoder,
    prototype_token=INP
).to(device)

# Eğitilmiş modeli yükle
model.load_state_dict(torch.load(model_path, map_location=device))
model.eval()
print("Model yüklendi!")

# === DÖNÜŞÜMLER ===
transform = transforms.Compose([
    transforms.Resize((252, 252)),  # Eğitimdeki boyutla aynı
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# === ANOMALİ HARİTASI FONKSİYONU ===
def generate_anomaly_map(img_path):
    """Görüntü için anomali haritası oluştur"""
    # Görüntüyü yükle ve ön işle
    image = Image.open(img_path).convert('RGB')
    input_tensor = transform(image).unsqueeze(0).to(device)

    # Modelden çıktı al
    with torch.no_grad():
        en, de, _ = model(input_tensor)
        anomaly_map, _ = cal_anomaly_maps(en, de, input_tensor.shape[-1])

        # Gaussian smoothing
        gaussian_kernel = get_gaussian_kernel(kernel_size=5, sigma=4).to(device)
        anomaly_map = gaussian_kernel(anomaly_map)

    # Anomali haritasını al
    anomaly_map_np = anomaly_map.squeeze().cpu().numpy()

    # Anomali haritasını normalize et (0-1 aralığına)
    anomaly_min = anomaly_map_np.min()
    anomaly_max = anomaly_map_np.max()

    if anomaly_max - anomaly_min < 1e-8:
        anomaly_map_norm = np.zeros_like(anomaly_map_np)
    else:
        anomaly_map_norm = (anomaly_map_np - anomaly_min) / (anomaly_max - anomaly_min)

    return anomaly_map_norm

# === TÜM DEFECT GÖRÜNTÜLERİNİ İŞLE ===
print("\n=== DEFECT GÖRÜNTÜLERİNİ İŞLEME ===")

# Tüm anomali skorları için liste (quantile hesaplaması için)
all_anomaly_values = []
all_anomaly_maps = []
image_paths = []

# Öncelikle tüm görüntüleri işle ve anomali haritalarını topla
for img_name in tqdm(sorted(os.listdir(defect_test_dir))):
    if not img_name.endswith(('.png', '.jpg')):
        continue

    img_path = os.path.join(defect_test_dir, img_name)

    # Anomali haritasını oluştur
    anomaly_map = generate_anomaly_map(img_path)

    # Tüm değerleri topla (quantile hesaplamak için)
    all_anomaly_values.extend(anomaly_map.flatten())

    # Haritayı ve görüntü yolunu sakla
    all_anomaly_maps.append(anomaly_map)
    image_paths.append(img_path)

# 0.95 quantile değerini hesapla
q95_threshold = np.quantile(all_anomaly_values, 0.95)
print(f"0.95 quantile eşik değeri: {q95_threshold:.4f}")

# Şimdi tüm görüntüler için sonuç görselleştirmelerini oluştur
for i, (img_path, anomaly_map) in enumerate(zip(image_paths, all_anomaly_maps)):
    img_name = os.path.basename(img_path)
    img_basename = os.path.splitext(img_name)[0]

    # Ground truth - Dosya adı aynı, sadece dizin farklı
    gt_path = os.path.join(gt_dir, f"{img_basename}.png")
    gt_mask = None
    if os.path.exists(gt_path):
        gt = cv2.imread(gt_path, cv2.IMREAD_GRAYSCALE)
        gt = cv2.resize(gt, (252, 252))
        gt_mask = (gt > 127).astype(np.uint8)
    else:
        print(f"Ground truth bulunamadı: {gt_path}")

    # Orijinal görüntü
    orig_img = np.array(Image.open(img_path).convert('RGB').resize((252, 252)))

    # Anomali maskesi (0.95 quantile kullanarak)
    binary_mask = (anomaly_map > q95_threshold).astype(np.uint8)

    # Morfolojik işlemler (gürültü temizleme)
    kernel = np.ones((3, 3), np.uint8)
    binary_mask = cv2.morphologyEx(binary_mask, cv2.MORPH_OPEN, kernel)
    binary_mask = cv2.morphologyEx(binary_mask, cv2.MORPH_CLOSE, kernel)

    # Overlay görüntüsü
    overlay = orig_img.copy()

    # Anomali konturları (yeşil)
    contours, _ = cv2.findContours(binary_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cv2.drawContours(overlay, contours, -1, (0, 255, 0), 2)

    # Ground truth konturları (kırmızı) - eğer varsa
    if gt_mask is not None:
        gt_contours, _ = cv2.findContours(gt_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        cv2.drawContours(overlay, gt_contours, -1, (255, 0, 0), 1)

    # Görselleştirme
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))

    # Orijinal görüntü
    axs[0, 0].imshow(orig_img)
    axs[0, 0].set_title(f"Orijinal - {img_name}")
    axs[0, 0].axis('off')

    # Anomali haritası
    im = axs[0, 1].imshow(anomaly_map, cmap='jet')
    axs[0, 1].set_title(f"Anomali Haritası")
    axs[0, 1].axis('off')
    fig.colorbar(im, ax=axs[0, 1], fraction=0.046, pad=0.04)

    # Ground truth (eğer varsa, yoksa boş bırak)
    if gt_mask is not None:
        axs[1, 0].imshow(gt_mask, cmap='gray')
        axs[1, 0].set_title("Ground Truth")
    else:
        axs[1, 0].imshow(np.zeros((252, 252)), cmap='gray')
        axs[1, 0].set_title("Ground Truth (Yok)")
    axs[1, 0].axis('off')

    # Overlay (model tahmini + ground truth)
    axs[1, 1].imshow(overlay)
    axs[1, 1].set_title(f"Model Tahmini (Yeşil) + Ground Truth (Kırmızı)")
    axs[1, 1].axis('off')

    plt.tight_layout()
    plt.savefig(os.path.join(output_dir, f"{img_basename}_comparison.jpg"), dpi=150)
    plt.close()

    # Ayrıca anomali haritası ve maskeyi de kaydet
    cv2.imwrite(os.path.join(output_dir, f"{img_basename}_heatmap.png"),
                (anomaly_map * 255).astype(np.uint8))
    cv2.imwrite(os.path.join(output_dir, f"{img_basename}_mask.png"),
                binary_mask * 255)

print(f"\n✅ Tüm işlemler tamamlandı! Sonuçlar {output_dir} klasörüne kaydedildi.")
print(f"Anomali tespiti için kullanılan eşik değeri (0.95 quantile): {q95_threshold:.4f}")
```
